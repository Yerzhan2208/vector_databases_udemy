```python
import pinecone
from pinecone import Pinecone, ServerlessSpec
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(), override=True)
```

```python
pc = Pinecone(api_key = os.environ.get("PINECONE_API_KEY"), environment = os.environ.get("PINECONE_ENV"))
```

```python
pc.list_indexes()
```

```python
index_name = "my-index"
dimension = 3
metric = "cosine"
```

```python
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
```

```python
pc.list_indexes()
```

```python
pc.create_index(
    name = index_name,
    dimension = dimension,
    metric = metric,
    spec = ServerlessSpec(
        cloud = "aws",
        region = "us-east-1"
    )
)
```

```python
pc.list_indexes()
```

```python
index_name_2 = "my-index-2"
dimension_2 = 1536
metric_2 = "euclidean"
```

```python
pc.create_index(
    name = index_name_2,
    dimension = dimension_2,
    metric = metric_2,
    spec = ServerlessSpec(
        cloud = "aws",
        region = "us-east-1"
    )
)
```

```python
pc.list_indexes()
```

```python
index = pc.Index(name = index_name)
```

```python
index.upsert([
    ("Dog", [4.,0.,1.]),
    ("Cat", [4.,0.,1.]),
    ("Chicken", [2.,2.,1.]),
    ("Mantis", [6.,2.,3.]),
    ("Elephant", [4.,0.,1.])
])
```
